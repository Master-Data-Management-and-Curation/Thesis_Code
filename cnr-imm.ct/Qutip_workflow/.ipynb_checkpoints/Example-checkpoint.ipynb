```python
#Let's define a simple spin system
#Hamiltonian
Ham="-\gamma_e B_0 S_z -\gamma_e B_1 \cos(\omega t) S_x"
#Constants
gamma_e = 2*np.pi*28e9 # [rad/(sT)]  (electron gyromagnetic ratio)
B_0=0.1 # Tesla
B_1=0.1 # Tesla
omega=2 * np.pi * 10e6 #rad/s

#Operators
S_z=qt.sigmaz()
S_x=qt.sigmax()

```

```python
#Eingenvalues in time

Bval=np.linspace(-B_0, B_0, 1000)
eigv = np.zeros([2, len(Bval)])

#Eigenvalues calculation for time independent Hamiltonian part
for i, B in enumerate(Bval):
    H=-gamma_e*B*S_z
    eigenvalues, eigenstates = H.eigenstates()
    eigv[0, i] = eigenvalues[0]
    eigv[1, i] = eigenvalues[1]



# Eigenvalues plot
fig1, ax1 = plt.subplots()
ax1.plot(Bval, eigv[0], label="First eigenvalue") 
ax1.plot(Bval, eigv[1], label="Second eigenvalue")
ax1.set_xlabel('B [T]')
ax1.set_ylabel('Energy [MHz]') 
ax1.legend(loc=(0.75, 0.45))
ax1.set_title('Eingenvalues vs B')
plt.show()
```

```python
#Time evolution
times=np.linspace(0, 1e-6, 1001)

#Complete Hamiltonian
H_ind=-gamma_e*B_0*S_z
H_time=-gamma_e*B_1*S_x

def time_dependence(t, args):
    omega = args['omega']
    value=np.cos(omega*t)
    return value

H_tot=[H_ind,[H_time,time_dependence]]

#State |+>
Psi0= qt.Qobj([[1], [0]]) 

#Time evolution
options_gate ={'nsteps': 5000, 'store_final_state': True}
result=qt.sesolve(H_tot,Psi0, times, e_ops=S_z, args={'omega': omega}, options=options_gate)

#Plots
fig_gate, axes_gate = plt.subplots(1, 1, figsize=(8, 5))
axes_gate.plot(result.times, result.expect[0], label=r'$\langle S_{z} \rangle$')
axes_gate.set_xlabel('Time (µs)')
axes_gate.set_ylabel('Expectation Value')
axes_gate.set_title('Time Evolution')
axes_gate.legend(loc='best')
axes_gate.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Gather Data


simulation_name = "Simple Spin Time Evolution Example"
output_filename = "example_output.qutip.json"

# --- Format sections using the library functions ---

# Program Info
prog_info = jw.format_program_info() # Uses QuTiP default

# System Info 
sys_info = jw.format_system_info(name="Single Spin-1/2 System", num_qubits=1)

# Hamiltonian Info
h_params_dict = {
    "gamma_e": (gamma_e, "rad/(s*T)"),
    "B_0": (B_0, "T"),
    "B_1": (B_1, "T"),
    "omega": (omega, "rad/s")
}
h_params_list = jw.format_hamiltonian_parameters(h_params_dict)

hamiltonian_dict = jw.format_spin_hamiltonian(
    name="Time-Dependent Spin Hamiltonian",
    formula=Ham, # Use the 'Ham' variable defined earlier
    parameters_list=h_params_list
)

# Operators
ops_to_save = {"Sz": S_z, "Sx": S_x}
operators_list = [jw.format_operator(name, op) for name, op in ops_to_save.items() if op is not None]


# States
states_to_save = {"Initial |+>": Psi0}
states_list = [jw.format_state(label, state, state_type_hint='ket') for label, state in states_to_save.items() if state is not None]


# Results
results_list = []
bval_var_dict = jw.format_variable_dict("Magnetic Field B", Bval, "T")
eig_result = jw.format_eigenvalue_result("Eigenvalues vs B", eigv, bval_var_dict)
results_list.append(eig_result)

# Time Evolution
time_var_dict = jw.format_variable_dict("Time", times, "s")
e_ops_names = ['<Sz>']
expect_vals = result.expect[0]

evo_result = jw.format_time_evolution_result(
    property_name="Expectation Value <Sz> vs Time",
    expect_values=expect_vals,
    e_ops_names=e_ops_names,
    time_variable_dict=time_var_dict,
    qutip_result_stats=result.stats,
    tlist=times,
    solver_options=options_gate
)
results_list.append(evo_result)


# --- Assemble and Save JSON ---
output_file = jw.create_simulation_json(
    filename=output_filename,
    program_info=prog_info,
    system_info=sys_info,
    hamiltonian_info=hamiltonian_dict,
    operators_list=operators_list,
    states_list=states_list,
    results_list=results_list,
    simulation_name=simulation_name
)

if output_file:
    print(f"JSON saved to: {output_file}")
else:
    print("JSON saving failed.")
```
